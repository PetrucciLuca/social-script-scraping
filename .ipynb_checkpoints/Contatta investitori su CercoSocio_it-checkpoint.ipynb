```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service as ChromeService
from selenium.webdriver.chrome.options import Options
import time
```

```python
def main():
    # Configurazione iniziale
    username = ''
    password = ''

    chrome_options = Options()
    chrome_options.add_argument('--no-default-browser-check')
    chrome_options.add_argument('--disable-default-apps')
    chrome_options.add_argument('--disable-infobars')
    chrome_options.add_argument('--disable-popup-blocking')
    chrome_options.add_argument("--disable-search-engine-choice-screen")

    driver_path = r'chromedriver-137.exe'
    service = ChromeService(driver_path)
    driver = webdriver.Chrome(service=service, options=chrome_options)
    driver.get("https://www.cercosocio.it")
    
    # Richiedi pagina iniziale
    start_page = int(input("Inserisci il numero della pagina da cui partire: "))
    base_url = "https://www.cercosocio.it/main.php?mod=search&q=1&pag="
    
    # Loop sulle pagine
    current_page = start_page
    while True:
        try:
            # Carica la pagina corrente
            url = f"{base_url}{current_page}"
            print(f"Aprendo la pagina: {url}")
            driver.get(url)
            time.sleep(3)  # Attendi il caricamento della pagina
            
            # Trova i link "Apri la scheda"
            elements = driver.find_elements(By.XPATH, "//a[contains(., 'Apri la scheda')]")
            links = [element.get_attribute('href') for element in elements]
            
            if not links:
                print(f"Nessun link trovato sulla pagina {current_page}. Terminato.")
                break
            
            # Itera sui link trovati
            for link in links:
                driver.get(link)
                time.sleep(5)  # Attendi il caricamento della pagina
                
                # Verifica se esiste il WebElement specificato
                try:
                    #driver.find_element(By.XPATH, "//strong[contains(text(), 'Mai contattato')]")
                    
                    # Esegui le azioni richieste
                    contact_button = driver.find_element(By.XPATH, "//a[contains(.,'Contatta ')]")
                    contact_button.click()
                    time.sleep(5)  # Attendi che si apra la pagina
                    
                    print("Invio Messaggio")
                    
                    # Inserisci oggetto
                    subject_field = driver.find_element(By.XPATH, "//input[@id='oggetto']")
                    subject_field.send_keys("Investi su BGSC")
                    
                    # Inserisci messaggio
                    message_field = driver.find_element(By.XPATH, "//textarea[@name='message']")
                    message_field.send_keys("""
                    Board Game Social Club è alla ricerca di soci finanziatori ...

                    Ad oggi, la nostra community conta oltre 1.200 utenti registrati, più di 200 GameMaster ...

                    Stiamo preparando il lancio della nuova applicazione ...

                    Cerchiamo uno o più soci finanziatori, ... 

                    Saluti
                    Luca Petrucci
                    www.boardgamesocialclub.com
                    Email: 
                    Telefono: 
                    """)
                    
                    # Invia messaggio
                    submit_button = driver.find_element(By.XPATH, "//input[@type='submit']")
                    submit_button.click()
                    time.sleep(3)  # Attendi il completamento
                    
                except Exception as e:
                    print(f"Elemento 'Mai contattato' non trovato per il link: {link}. Passo al successivo.")
                    continue
            
            # Passa alla pagina successiva
            current_page += 1
        
        except Exception as e:
            print(f"Errore durante l'elaborazione della pagina {current_page}: {e}")
            break

    # Chiudi il browser
    driver.quit()
```

```python
if __name__ == "__main__":
    main()
```
